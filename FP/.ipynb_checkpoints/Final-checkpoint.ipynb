## Import Data

```python
# import statements
import pandas as pd
import numpy as np
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as scs
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder
```

```python
# data set train.csv
train="data/train.csv"
```

```python
# read data into pandas
data=pd.read_csv(train, index_col=0)
```

```python
# put data in dataframe
df = pd.DataFrame(data)
```

## Run preliminary correlation analysis

```python
# calculate a correlation matrix for the data
corr = df.corr()
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(50, 50))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(240, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin=-.5, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt="f")
```

```python
# divide categorical variables into nominal, ordinal
nominal_encode = ['MSSubClass', 'MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities', 'LotConfig',
            'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType', 'HouseStyle', 
            'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Foundation', 'Heating', 'CentralAir',
            'Electrical', 'Functional', 'GarageType', 'PavedDrive', 'MiscFeature', 'MoSold', 'YrSold',
            'SaleType', 'SaleCondition', 'Fence']

ordinal_encode_standard = ['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond',
                 'HeatingQC', 'KitchenQual', 'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC']

ordinal_encode_other = ['BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'GarageFinish']

binning = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']
```

```python
pp = sns.pairplot(data=data,
                  y_vars=['SalePrice'],
                  x_vars=nominal_encode)
```

```python
pp = sns.pairplot(data=data,
                  y_vars=['SalePrice'],
                  x_vars=ordinal_encode_standard)
```

```python
# one hot encoding
for col in nominal_encode:
    df = pd.concat([df,pd.get_dummies(df[col], prefix=col)],axis=1)
    df.drop([col],axis=1, inplace=True)
```

```python
mask_1 = {'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0}
mask_2 = {'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0}
mask_3 = {'GLQ':6, 'ALQ':5, 'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0}
mask_4 = {'Fin':3, 'RFn':2, 'Unf':1, 'NA':0}
```

```python
# ordinal encoding for standard codes
for col in ordinal_encode_standard:
    df[col].replace(mask_1, inplace=True)
```

```python
# ordinal encoding for special codes
df['BsmtExposure'].replace(mask_2, inplace=True)
df['BsmtFinType1'].replace(mask_3, inplace=True)
df['BsmtFinType2'].replace(mask_3, inplace=True)
df['GarageFinish'].replace(mask_4, inplace=True)
```

```python
y = df['SalePrice'].fillna(0).to_numpy()
x = df.drop('SalePrice', axis=1).fillna(0).to_numpy()
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```
